```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
df =pd.read_csv('../data/house_price_datasets.csv')
print(df.head())
```

```python
# fig , axes =plt.subplots(1,3 ,figsize=(16,10))

# Price Distribution
sns.histplot(df['Price'],kde=True , bins=40 )
plt.title("Price Distribution")
plt.savefig('Price_Distribution.png')
plt.show()

# Correlation Matrix
ys = {'fontsize':15 ,  'color':'black'}
sns.heatmap(df.corr() , annot=True ,annot_kws=ys , fmt='.1f' , cmap='coolwarm' , linewidths=10 )
plt.title("Correlation Matrix")
plt.savefig('Correlation.png')
plt.show()

# Area Vs Price
sns.scatterplot(x=df['Area_sqr'],y=df['Price'] , alpha=0.5 , s=10)
plt.title("Area Vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.savefig("Area_Vs_Price.png")
plt.show()
```
